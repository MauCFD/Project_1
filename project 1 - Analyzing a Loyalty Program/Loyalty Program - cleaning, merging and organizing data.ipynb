```python
# Dependencies
import pandas as pd
import os
import numpy as np
```

```python
# Study data files
location_path = "data/CT_DB_Location.csv"
points_path = "data/CT_DB_Points.csv"
sales_path = "data/CT_DB_Sales.csv"

# Read all the csv files
location_df = pd.read_csv(location_path)
points_df = pd.read_csv(points_path)
sales_df = pd.read_csv(sales_path)

# Combine the data into a single dataset
combined_data = pd.merge(sales_df, points_df, on=["Client ID", "Quarter", "Month", "Year"], how="outer")
combined_data = pd.merge(combined_data, location_df, on= "Client ID", how="outer")

# Display the data table for preview
combined_data
```

```python
# Adding the Total Sales and Points, and Location columns
combined_data["Total Sales"] = combined_data["A"] + combined_data["B"] + combined_data["C"]
combined_data["Total Points"] = combined_data["Monthly Points"] + combined_data["Quarter Bonus 1"] + combined_data["Quarter Bonus 2"] + combined_data["Yearly Bonus 1"] + combined_data["Yearly Bonus 1"] 

# We know that the program estimates each point is worth $0.86 MXN, so we will create a column to know the Points worth in $
combined_data["Points Worth ($)"] = combined_data["Total Points"] * 0.86

# We would like to know the compliance of the client in their monthly sales, so we will add a column that calculates it.
combined_data["Compliance"] = round(combined_data["Total Sales"] / combined_data["Min Monthly Sales"],2)

# Reorganizing the data frame
Project_df = combined_data[["Client ID", "Town", "State", "Month", "Quarter", "Year", "Category", 
                               "Distributor", "Min Monthly Sales", "Total Sales", "Compliance", "Total Points", "Points Worth ($)"]]
Project_df['Total Sales'] = Project_df['Total Sales'].astype(np.int64)
Project_df['Total Points'] = Project_df['Total Points'].astype(np.int64)

# Organized and clean data frame we will be working with
Project_df
```

```python
# Export the Project_df into a csv
Project_df.to_csv("Project_df.csv",encoding="utf-8",index=False)
```
