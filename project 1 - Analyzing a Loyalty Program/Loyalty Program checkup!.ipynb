```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import os
import scipy.stats as st
import numpy as np
```

```python
# Study data files
location_path = "data/CT_DB_Location.csv"
points_path = "data/CT_DB_Points.csv"
sales_path = "data/CT_DB_Sales.csv"

# Read all the csv files
location_df = pd.read_csv(location_path)
points_df = pd.read_csv(points_path)
sales_df = pd.read_csv(sales_path)

# Combine the data into a single dataset
combined_data = pd.merge(sales_df, points_df, on=["Client ID", "Quarter", "Month", "Year"], how="outer")
combined_data = pd.merge(combined_data, location_df, on= "Client ID", how="outer")

# Display the data table for preview
combined_data
```

```python
# Adding the Total Sales and reorganizing the data frame
combined_data["Total Sales"] = combined_data["A"] + combined_data["B"] + combined_data["C"]
combined_data["Total Points"] = combined_data["Monthly Points"] + combined_data["Quarter Bonus 1"] + combined_data["Quarter Bonus 2"] + combined_data["Yearly Bonus 1"] + combined_data["Yearly Bonus 1"] 
    

Project_df = combined_data[["Client ID", "Town", "State", "Month", "Quarter", "Year", "Category", 
                               "Distributor", "Total Sales", "Total Points"]]
Project_df['Total Sales'] = Project_df['Total Sales'].astype(np.int64)
Project_df['Total Points'] = Project_df['Total Points'].astype(np.int64)

# Data frame we will be working with
Project_df
```

```python
# Displaying the total categories
categories = Project_df.drop_duplicates("Category").sort_values("Category", ascending= True)
categories["Category"]
```

```python
# Displaying the total distributors
distributors = Project_df.drop_duplicates("Distributor").sort_values("Distributor", ascending= True)
distributors["Distributor"]
```

```python
# Displaying total sales by year
total_sales_by_year = Project_df.groupby(Project_df["Year"])["Total Sales"].sum()
total_sales_by_year
```

```python
# Displaying total sales by year and state
sales_by_state = Project_df.groupby(['State','Year'])["Total Sales"].sum()
sales_by_state
```
